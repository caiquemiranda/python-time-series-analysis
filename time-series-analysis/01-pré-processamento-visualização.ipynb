```python
# importando os dados terremotos na italia entre 2016-08-24 e 2016-11-30

df = pd.read_csv('./data/italy_earthquakes_from_2016-08-24_to_2016-11-30.csv').set_index('Time')
```

```python
# visualizando os primeiros registros

df.head()
```

```python
# visualizando os tipos de dados

df.dtypes
```

```python
# convertendo os dados do indice para datetime

df.index = pd.to_datetime(df.index)
```

```python
# descrição estatística dos dados

df.describe()
```

```python
# visualização grafica dos dados

# magnitude média por dia
df['Magnitude'].resample('D').apply([np.mean]).plot()
plt.title('Magnitude média por dia')
plt.ylabel('Magnitude média')
plt.show()
```

```python
# magnitude média a cada 2 dias

df['Magnitude'].resample('2D').apply([np.mean]).plot()
plt.title('Magnitude média a cada 2 dias')
plt.ylabel('Magnitude média')
plt.show()
```

```python
# magnitude maior que 3

mag3 = df.loc[df['Magnitude'] >= 3.0, 'Magnitude']
mag3.head()
```

```python
# média móvel de magnitude maior que 3

pd.Series.rolling(mag3, window=25).mean().plot(style='-g')
plt.title('Media movel da magnitude maior que 3')
plt.xlabel('Tempo')
plt.ylabel('Magnitude')
plt.show();
```

```python
# importando dados

data = pd.read_csv('./data/avocado.csv', parse_dates=['Date'], index_col='Date')
```

```python
# visualizando os primeiros registros

data.head()
```

```python
# agrupamento dos dados por dia e visualização

summer = data.groupby('Date')['Total Volume'].mean().reset_index()
summer.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,6))
summer.set_index('Date').plot(ax=ax, marker='o', linestyle='--', color='red', linewidth=2)
plt.title('Total Volume por dia')
plt.ylabel('Total Volume')
plt.show();
```

```python
# metricas dos dados

fig, ax = plt.subplots(1, 1, figsize=(12,10))
(summer.set_index('Date')
    .assign(month=lambda df: df.index.month)
    .groupby('month')['Total Volume'].agg(['mean', 'std', 'min', 'max', 'median'])
    .plot(ax=ax, marker='o', linewidth=3)
    )
ax.set_xlabel('Mes')
ax.set_ylabel('Total Volume')
plt.title('Metricas dos dados')
plt.show();
```

```python
# importando dados de uma série temporal de ações

df_stocks = pd.read_csv('./data/all-stocks-2006-01-01-to-2018-01-01.csv', parse_dates=['Date'])
```

```python
# visualizando os primeiros registros

df_stocks.head()
```

```python
# definindo um range de datas com a frequencia de D

rng = pd.date_range(start='2006-01-01', end='2018-01-01', freq='B')

# não pegar datas repetidas
rng[~rng.isin(df_stocks.Date.unique())]
```

```python
# selecionando a ultima data de cada ano

sample_dates = pd.date_range(start='2006-01-01', end='2018-01-01', freq='B')

year_end_dates = sample_dates[sample_dates.is_year_end]
year_end_dates
```

```python
# 5 piores desempenhos das ações

worst_stocks = df_stocks[df_stocks.Date == df_stocks.Date.max()].sort_values('Open').head(5)
worst_stocks
```

```python
# nome das 5 piores ações

ws = worst_stocks.Name.values
ws
```

```python
# estructura dos dados

tdf = df_stocks.copy()
tdf = df_stocks.set_index('Date')
```

```python
# grafico das 5 piores ações

tdf[tdf.Name.isin(ws)].groupby('Name').Open.plot(title='Tendencia Historica das 5 piores ações de 2017 na abertura', legend=True);
```

```python
# criando grafico de tendencia historica das 5 melhores ações

best_stocks = df_stocks[df_stocks.Date == df_stocks.Date.max()].sort_values('Open', ascending=False).head(5)
best_stocks
```

```python
# nome das 5 melhores ações

bs = best_stocks.Name.values
bs
```

```python
# grafico das 5 melhores ações

tdf[tdf.Name.isin(bs)].groupby('Name').Open.plot(title='Tendencia Historica das 5 melhores ações de 2017 na abertura', legend=True);
```

```python
# importando dados

df = pd.read_csv('./data/italy_earthquakes_from_2016-08-24_to_2016-11-30.csv').set_index('Time')
```

```python
# visualizando os primeiros registros

df.head()
```

```python
# tipos de dados

df.dtypes
```

```python
# convertendo os dados do indice para datetime(Time->datetime)

df.index = pd.to_datetime(df.index)
```

```python
# visualizando após a transformação

df
```

```python
# filtro de profundidade entre 0 e 22 

plt.figure(figsize=(12,6))

depth = df[(df['Depth/Km'] < 22) & (df['Depth/Km'] >= 0)]['Depth/Km']
depth.plot(kind='hist', 
           bins=50, 
           title='Histograma da profundidade dos terremotos', 
           legend=True)
plt.xlabel('Profundidade')
plt.ylabel('Frequencia')
plt.show();
```

```python
# exibindo variações de series temporais
```

```python
# visualizando os primeiros registros

df.index = pd.to_datetime(df.index)
df.head()
```

```python
# filtro dos dados em profundidade maior ou igual a 3.0 junto com a magnitude

depth_magn = df.where((df['Magnitude'] >= 3.0)).dropna()[['Magnitude', 'Depth/Km']]
dm = depth_magn.groupby(depth_magn.index.hour).mean()
dm.head()
```

```python
# contruindo o grafico
fig = plt.figure()

ax = fig.add_subplot(111)
ax.set_ylim([2.5, 4.0])
ax.set_ylabel('Magnitude')
ax.set_xlabel('Hora do dia')
ax.yaxis.label.set_color('blue')

ax2 = ax.twinx()
ax2.set_ylim([4.0, 12.0])
ax2.set_ylabel('Profundidade/Km')
ax2.set_xlabel('Hora do dia')
ax2.yaxis.label.set_color('red')

# plotando o grafico
dm['Magnitude'].plot(kind='bar', 
                     ax=ax, 
                     color='blue', 
                     width=0.5,
                     position=1)

dm['Depth/Km'].plot(kind='bar', 
                     ax=ax2, 
                     color='red', 
                     width=0.5,
                     position=0)

plt.grid(False)
plt.title('Magnitude e Profundidade durante o dia')
plt.show();
```
