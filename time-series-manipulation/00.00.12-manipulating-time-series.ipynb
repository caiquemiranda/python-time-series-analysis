```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# realação entre time series 
```

```python
stock_prices = pd.read_csv('./data/stock_prices.csv', index_col='Date', parse_dates = ['Date'])
stock_prices.head(3)
```

```python
stock_prices.dropna().info()
```

```python
# correlação entre os ativos

daily_returns = stock_prices.pct_change()

sns.jointplot(x='BVSP', y='AMZN', data=daily_returns);
```

```python
# calculo de todos os ativos
```

```python
correlations = daily_returns.corr()
correlations
```

```python
plt.figure(figsize=(12,6))
plt.title('Correlation Matrix')
sns.heatmap(correlations, annot=True, fmt='.2f', cmap='coolwarm');
```
